# Cleaning process

1. Extract important features from the corpus or name.
2. Filling up NaN values manually by searching up the web.

```python
import re

def extract_brand(text):
    return text.split()[0]

def extract_ram(text):
    match = re.search(r'RAM(\d+)', text)
    if match:
        return int(match.group(1))
    return None

def extract_system(text):
    match = re.search(r'System(.*?)Processor', text)
    if match:
        return match.group(1).strip()
    return None

def extract_processor(text):
    match = re.search(r'Processor (.*?) ', text)
    if match:
        return match.group(1).strip()
    return None

def extract_storage(text):
    match = re.search(r'Storage(\d+)', text)
    if match:
        return int(match.group(1))
    return None
```

```python
Cellphone_data['Brand'] = Cellphone_data['name'].apply(extract_brand)
Cellphone_data['RAM'] = Cellphone_data['corpus'].apply(extract_ram)
Cellphone_data['System'] = Cellphone_data['corpus'].apply(extract_system)
Cellphone_data['Processor'] = Cellphone_data['corpus'].apply(extract_processor)
Cellphone_data['Storage'] = Cellphone_data['corpus'].apply(extract_storage)
```

```python
unique_systems = Cellphone_data['System'].unique()
unique_processors = Cellphone_data['Processor'].unique()
unique_systems, unique_processors
```

```python
def clean_processor(processor):
    if pd.isnull(processor):
        return None
    if 'mediatek' in processor.lower():
        return 'MediaTek'
    if 'qualcomm' in processor.lower():
        return 'Qualcomm'
    if 'apple' in processor.lower():
        return 'Apple'
    return 'Other'

```

```python
Cellphone_data['Processor'] = Cellphone_data['Processor'].apply(clean_processor)
```

```python
def clean_system(system):
    if pd.isnull(system):
        return None
    if 'android' in system.lower():
        return 'Android'
    if 'ios' in system.lower():
        return 'iOS'
    if 'tizen' in system.lower():
        return 'Tizen'
    return system

Cellphone_data['System'] = Cellphone_data['System'].apply(clean_system)
```

# Recommendation Function

```python
def recommend(brand = None, system = None, min_price = None, max_price = None, top_n = 10):
    filtered_data = Cellphone_data[
        (Cellphone_data['Brand'] == brand) &
        (Cellphone_data['System'] == system) &
        (Cellphone_data['price'] >= min_price) &
        (Cellphone_data['price'] <= max_price)
    ]
    
    if filtered_data.empty:
        filtered_data = Cellphone_data[
            (Cellphone_data['Brand'] == brand) &
            (Cellphone_data['price'] >= min_price) &
            (Cellphone_data['price'] <= max_price)
        ]
    
    if filtered_data.empty:
        filtered_data = Cellphone_data[
            (Cellphone_data['price'] >= min_price) &
            (Cellphone_data['price'] <= max_price)
        ]
        
    if filtered_data.empty:
        filtered_data = Cellphone_data[
            (Cellphone_data['Brand'] == brand)
        ]
    
    if filtered_data.empty:
        filtered_data = Cellphone_data[
            (Cellphone_data['System'] == system)
        ]
    
    if filtered_data.empty:
        print('No preferences listed!')
        return
    
    recommendations = filtered_data.sort_values(by = 'ratings', ascending = False).head(top_n)
    return list(recommendations.name.values)
```
